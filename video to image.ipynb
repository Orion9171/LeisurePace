```python
import cv2
import os

def get_images_from_video(video_name, time_F, output_folder):
    video_images = []
    vc = cv2.VideoCapture(video_name)
    c = 1

    if vc.isOpened():
        rval, video_frame = vc.read()
    else:
        rval = False

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    while rval:
        rval, video_frame = vc.read()

        if (c % time_F == 0):
            image_path = os.path.join(output_folder, f'image{c // time_F:04d}.jpg')
            cv2.imwrite(image_path, video_frame)
            video_images.append(image_path)
            print(f'Saved: {image_path}')  # Add this line to print the saved image paths
        c = c + 1

    vc.release()

    return video_images

time_F = 30
# Provide the full path to the video file if it's not in the script's directory
video_name = 'run1.mp4'
output_folder = 'output_images1'
video_images = get_images_from_video(video_name, time_F, output_folder)

for image_path in video_images:
    print(image_path)

print('Images saved to', output_folder)
```
